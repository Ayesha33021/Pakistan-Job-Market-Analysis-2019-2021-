```python
def clean_description(text):
    # Handle missing or non-string values safely
    if not isinstance(text, str):
        return []
    
    # Lowercase, tokenize, and filter punctuation/stopwords
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalnum() and word not in stop_words]

# Apply the function to create a new column of cleaned word lists
df['cleaned_words'] = df['JD'].apply(clean_description)
```

```python
def extract_skills_from_jd(text):
    if not isinstance(text, str):
        return []
    
    # Lowercase the entire text block
    text_lower = text.lower()
    
    found_skills = []
    
    # Loop over our dictionary to look for explicit skill matches
    for skill_key, clean_name in SKILL_DICTIONARY.items():
        # Escape special characters like ++ or # safely
        escaped_skill = re.escape(skill_key)
        
        # Use regex word boundaries (\b) to avoid matching 'go' inside 'google' or 'government'
        # For c++ and c#, we adjust boundaries because \b doesn't treat '+' or '#' as word characters
        if '++' in skill_key or '#' in skill_key:
            pattern = rf'\b{escaped_skill}(?:\s|\b|$)'
        else:
            pattern = rf'\b{escaped_skill}\b'
            
        if re.search(pattern, text_lower):
            found_skills.append(clean_name)
            
    return found_skills

# Apply the extractor directly to your JD column
df['extracted_skills'] = df['JD'].apply(extract_skills_from_jd)
```

```python
# Flatten all extracted skills across all rows
all_skills = [skill for skill_list in df['extracted_skills'] for skill in skill_list]

# Count frequencies
skill_counts = Counter(all_skills)

# Print out your true corporate demand metrics
print("--- TOP SKILLS REQUIRED BY EMPLOYERS ---")
for skill, count in skill_counts.most_common(15):
    print(f"{skill}: Found in {count} job descriptions")
```

```python
# 1. Extract the top 15 most common skills from your Counter
top_skills = skill_counts.most_common(15)

# 2. Unpack the data into separate lists for plotting
# We reverse them ([::-1]) so the highest-demanded skill sits proudly at the top
skills, counts = zip(*top_skills)
skills = list(skills)[::-1]
counts = list(counts)[::-1]

# 3. Create the plot
plt.figure(figsize=(10, 6))
bars = plt.barh(skills, counts, color='#2b5c8f', edgecolor='none', height=0.6)

# 4. Styling and Polish
plt.title('What Skills Do Employers Require?\n(Top Tech Skills Extracted from Job Descriptions)', 
          fontsize=14, fontweight='bold', pad=15, loc='left')
plt.xlabel('Number of Job Postings', fontsize=11, labelpad=10)
plt.ylabel('Technical Skills', fontsize=11)

# Remove the box border lines (spines) for a cleaner, modern corporate feel
for spine in ['top', 'right', 'bottom', 'left']:
    plt.gca().spines[spine].set_visible(False)

# Add faint vertical grid lines to easily track counts
plt.grid(axis='x', linestyle='--', alpha=0.5)

# 5. Add direct value labels to the end of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + (max(counts) * 0.01),       # Position slightly to the right of the bar end
             bar.get_y() + bar.get_height()/2,    # Vertically centered inside the bar row
             f'{int(width):,}',                  # Format with thousands separators if large
             va='center', ha='left', fontsize=10, fontweight='semibold', color='#333333')

# Prevent labels from getting chopped off at the edges
plt.tight_layout()

# Display the final chart
plt.show()
plt.savefig('Top Tech Skills Extracted from Job Descriptions.png', dpi=300, bbox_inches='tight')
```
